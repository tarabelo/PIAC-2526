## **Introducci칩n a los problemas de optimizaci칩n**

Muchos problemas relevantes en ciencia, ingenier칤a, econom칤a o inteligencia artificial pueden expresarse como **problemas de optimizaci칩n**. En t칠rminos generales, un problema de optimizaci칩n consiste en encontrar los valores de ciertas variables que **minimizan** o **maximizan** una **funci칩n objetivo**, respetando una serie de **restricciones**.

Formalmente, un problema de optimizaci칩n puede escribirse como:

$$
\begin{aligned}
\text{Minimizar (o maximizar)} \quad & f(x_1, x_2, \dots, x_n) \\
\text{sujeto a} \quad & g_i(x_1, \dots, x_n) \leq 0, \quad i = 1, \dots, m \\
& h_j(x_1, \dots, x_n) = 0, \quad j = 1, \dots, p
\end{aligned}
$$

donde:
- $f$ es la funci칩n objetivo,
- $g_i$ son restricciones de desigualdad,
- $h_j$ son restricciones de igualdad,
- y las variables $x_k$ pueden ser reales, enteras, binarias o discretas.

### Ejemplos cl치sicos de problemas de optimizaci칩n

- **Problema del viajante (TSP)**: Dado un conjunto de ciudades y las distancias entre ellas, encontrar el camino m치s corto que visite cada ciudad exactamente una vez y regrese al punto de origen.

- **Problema de la mochila (Knapsack)**: Dados objetos con un peso y un valor, seleccionar un subconjunto cuyo peso total no exceda una capacidad dada y cuyo valor total sea m치ximo.

- **Coloraci칩n de grafos**: Asignar colores a los nodos de un grafo de forma que nodos adyacentes no compartan color y se minimice el n칰mero total de colores utilizados.

- **Problema de satisfacibilidad booleana (SAT)**: Determinar si existe una asignaci칩n de valores booleanos a un conjunto de variables que satisfaga una f칩rmula l칩gica.

- **Programaci칩n lineal binaria/entera**: Optimizar una funci칩n lineal sujeta a restricciones lineales, donde las variables s칩lo pueden tomar valores/enteros.

Muchos de estos problemas son [**NP-duros**](https://es.wikipedia.org/wiki/NP-hard), es decir, su complejidad crece exponencialmente con el tama침o del problema.

<summary>Resumen sobre las clases de complejidad computacional</summary>

Cuando hablamos de la **complejidad computacional** de un problema, nos referimos a la cantidad de recursos (tiempo y/o espacio) necesarios para resolverlo en funci칩n del tama침o de la entrada. Esto se formaliza mediante **clases de complejidad**, que agrupan problemas seg칰n los recursos requeridos por sus algoritmos.

### Clases principales

#### 游댳 P (Polynomial Time)
- Conjunto de problemas que pueden resolverse en **tiempo polin칩mico** por una m치quina determinista.
- Es decir, existen algoritmos eficientes que resuelven el problema con un coste $O(n^k)$, para alguna constante $k$.
- **Ejemplo**: ordenamiento de una lista, b칰squeda en un grafo no ponderado.

#### 游댳 NP (Nondeterministic Polynomial Time)
- Problemas cuya soluci칩n puede **verificarse** en tiempo polin칩mico, aunque no sepamos c칩mo **encontrarla** eficientemente.
- Incluye todos los problemas de P (es decir, $\text{P} \subseteq \text{NP}$).
- **Ejemplo**: SAT, TSP, coloreado de grafos, problema de la mochila.

#### 游댳 NP-completo
- Subconjunto de NP que incluye los problemas **m치s dif칤ciles** de la clase.
- Un problema es NP-completo si:
  1. Pertenece a NP.
  2. Todo problema en NP puede reducirse a 칠l en tiempo polin칩mico.
- Si se resuelve uno en tiempo polin칩mico, **todos los problemas de NP tambi칠n se pueden resolver eficientemente**.
- **Ejemplo**: SAT (el primer problema demostrado como NP-completo), 3-COLOR, TSP en su versi칩n de decisi칩n (determinar si existe un camino menor o igual que un valor dado).

#### 游댳 NP-duro (NP-hard)
- Problemas **al menos tan dif칤ciles** como los NP-completos, pero **no necesariamente pertenecen a NP** (por ejemplo, pueden no tener soluciones verificables en tiempo polin칩mico).
- Suelen ser problemas de optimizaci칩n o de decisi칩n m치s generales.
- **Ejemplo**: versi칩n de optimizaci칩n del TSP (encontrar el camino m치s corto), QUBO, etc.

### 쯇 = NP?

La pregunta de si $\text{P} = \text{NP}$ es uno de los problemas abiertos m치s importantes de la inform치tica te칩rica. Si se demuestra que $\text{P} = \text{NP}$, todos los problemas cuya soluci칩n se puede verificar eficientemente tambi칠n podr칤an **resolverse** eficientemente, lo que tendr칤a enormes implicaciones en criptograf칤a, planificaci칩n, dise침o autom치tico, etc.

### Clases de complejidad cu치ntica

#### 游댳 BQP (Bounded-error Quantum Polynomial Time)
- Problemas que pueden resolverse en **tiempo polin칩mico por una computadora cu치ntica** con un error de probabilidad acotado (< 1/3).
- Es la **contraparte cu치ntica de P**, e incluye algunos problemas que no se sabe c칩mo resolver eficientemente con algoritmos cl치sicos.
- **Ejemplo**: factorizaci칩n de enteros (algoritmo de Shor), logaritmo discreto, simulaci칩n de sistemas cu치nticos.

$$
\text{P} \subseteq \text{BPP} \subseteq \text{BQP}
$$

Donde **BPP** (Bounded-error Probabilistic Polynomial time) representa problemas que pueden resolverse eficientemente con algoritmos probabil칤sticos cl치sicos.

#### 游댳 QMA (Quantum Merlin-Arthur)
- Contraparte cu치ntica de **NP**: problemas para los que un verificador cu치ntico puede comprobar una prueba cu치ntica en tiempo polin칩mico.
- Existen problemas QMA-completos, como ciertas versiones del **problema del Hamiltoniano local**.

---

### Relaciones generales entre clases (parcialmente conocidas)

- Se sabe que:

$$
\text{P} \subseteq \text{NP}, \quad \text{P} \subseteq \text{BQP}
$$

- Pero **no se conoce** si:

$$
\text{NP} \subseteq \text{BQP} \quad \text{o si} \quad \text{BQP} \subseteq \text{NP}
$$

---

### Relevancia para la computaci칩n cu치ntica

Problemas como **QUBO** o el **modelo de Ising** pertenecen a la clase **NP-hard** y, en general, **no se espera** que puedan resolverse eficientemente en computadoras cl치sicas. Sin embargo, ciertos algoritmos cu치nticos como el **Quantum Annealing** (usado en D-Wave) o el **QAOA (Quantum Approximate Optimization Algorithm)** ofrecen aproximaciones prometedoras, que caen dentro (o cerca) de la clase BQP o heur칤sticas asociadas.

## **Problemas de optimizaci칩n binaria cuadr치tica sin restricciones (QUBO)**

Un tipo de problemas en los que se est치 usando la computaci칩n cu치ntica son los denominados [QUBO](https://en.wikipedia.org/wiki/Quadratic_unconstrained_binary_optimization) (_Quadratic Unconstrained Binary Optimization_).

Este tipo de problemas consisten en minimizar una funci칩n $f_Q(x)$ con la siguiente forma:

$$
f_Q(x) = x^TQx = \sum_{i=0}^{n-1}\sum_{j=0}^i q_{ij}x_ix_j
$$

donde $Q \in \mathbb{R}^{n\times n}$ es una matriz triangular (o sim칠trica) y $x$ un vector de componentes binarias ($x_i \in \{0,1\}$).

Los problemas QUBO son [NP-duros](https://es.wikipedia.org/wiki/NP-hard).

Muchos modelos de optimizaci칩n combinatoria pueden expresarse como problemas QUBO, por ejemplo modelos de programaci칩n lineal o problemas como el coloreado de grafos o el problema del viajante.

### Ejemplo: algoritmo MAX-CUT

Sea $G = (V, E)$ un grafo pesado no dirigido con $n$-nodos y pesos $w_{ij}>0$, $w_{ij}=w_{ji}$, con $(i,j)\in E$ y $w_{ij}=0$ si $(i,j)\notin E$.

Objetivo: dividir el grafo en dos conjuntos tal que la suma de los pesos de las aristas entre ambos conjuntos sea m치ximo.


<center><img src="https://drive.google.com/uc?export=view&id=1t37PWgnsVWRkX16wcbmFSPZNZ1D0dUOS" alt="Ejemplo MAX-CUT" width="700"  /></center>

En este grafo, los nodos $0$ y $3$ se asignan a un conjunto y $1$, $2$ y $4$ al otro. El peso del corte es $19$.

El algoritmo procede asignando a cada v칠rtice un valor $x_i = \{0,1\}$ de forma que el grafo queda dividido en dos conjuntos: los v칠rtices con $x_i = 0$ y aquellos con $x_i = 1$.

El algoritmo MAX-CUT busca el n칰mero binario $\textbf{x}=x_0\cdots x_{n-1}$ que maximice la funci칩n de coste:

$$
C(\textbf{x}) = \sum_{i,j = 0}^{n-1} w_{ij} x_i (1-x_j)
$$

siendo $w_{i,j} \ne 0$ solo si $(i,j) \in E$.

Si los v칠rtices $i$ y $j$ est치n en el mismo conjunto: $x_i = x_j \Rightarrow x_i (1-x_j) = 0$.

Si los v칠rtices $i$ y $j$ est치n en diferentes conjunto: $x_i \ne x_j \Rightarrow x_i (1-x_j) = 1$ 칩 $x_j (1-x_i) = 1$.

En la soluci칩n del grafo de la imagen $x_0=x_3=0$ y $x_1=x_2=x_4=1$, y el peso es:

$$
C(01101) = w_{01}x_0(1-x_1)+ w_{10}x_1(1-x_0)+w_{12}x_1(1-x_2)+w_{21}x_2(1-x_1)+\\
w_{13}x_1(1-x_3)+w_{31}x_3(1-x_1)+w_{23}x_2(1-x_3)+w_{32}x_3(1-x_2)+\\
w_{24}x_2(1-x_4)+w_{42}x_4(1-x_2)+w_{34}x_3(1-x_4)+w_{43}x_4(1-x_3)=\\
w_{10}+w_{13}+w_{23}+w_{43} = 3+4+7+5 = 19
$$

As칤, $C(\textbf{x})$ es la suma de los pesos de las aristas que separan ambos conjuntos.

---

#### MAXCUT como problema QUBO
Los valores $w_{ij}$ son $\ne 0$ solo para las aristas del grafo.

Para una arista $(i,j)\in E$, los t칠rminos en $C(x)$ son:

$$
x_i(1-x_j) + x_j(1-x_i) = x_i+x_j-2x_ix_j
$$

Por tanto, podemos escribir la funci칩n de coste como:

$$
C(x) = \sum_{(i,j) \in E} w_{ij} (x_i + x_j - 2 x_i x_j)
$$

Para expresarlo como un **problema QUBO**, que requiere una **minimizaci칩n**, usamos el negativo:

$$
f(x) = -\sum_{(i,j) \in E} w_{ij} (x_i + x_j - 2 x_i x_j)
$$



#### Matriz $Q$ del QUBO

Reorganizamos la expresi칩n en forma cuadr치tica:

$$
f(x) = x^T Q x
$$

Donde $Q \in \mathbb{R}^{n \times n}$ es una matriz sim칠trica definida como:

$$
q_{ij} =
\begin{cases}
- \sum_{k \ne i} w_{ik} & \text{si } i = j \\
2 w_{ij} & \text{si } (i,j) \in E \text{ y } i \gt j \\
0 & \text{si } (i,j) \notin E
\end{cases}
$$

---

#### Ejemplo: grafo anterior como QUBO

La matriz $Q$ correspondiente es:

$$
Q = \begin{pmatrix}
-3 & 0 & 0 & 0 & 0 \\
6 & -9 & 0 & 0 & 0 \\
0 & 4 & -10  & 0 & 0 \\
0 & 8 & 14  & -16 & 0 \\
0 & 0 & 2  & 10 & -6
\end{pmatrix}
$$

Se puede comprobar que:

$$
f(01101) = x^T Q x = -19
$$

#### Ejemplo: programa Python que genera el grafo de la imagen, obtiene Q y resuelve el MAXCUT por fuerza bruta.

```python
import numpy as np
import networkx as nx  # Librer칤a para manejar grafos

# Ejemplo de grafo con 5 nodos
nnodes = 5
G = nx.Graph()
# A침ade nodos y aristas
G.add_nodes_from(np.arange(0,nnodes,1))
edges = [(0,1,3),(1,2,2),(1,3,4),(2,4,1),(2,3,7),(3,4,5)]
G.add_weighted_edges_from(edges)

# Mostramos el grafo
layout = nx.random_layout(G,seed=10)
colors = ['red', 'green', 'lightblue', 'yellow', 'magenta']
nx.draw_networkx(G, layout, node_color=colors)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos=layout, edge_labels=labels);
```

```python
# Funci칩n para obtener la matriz Q
def maxcut_qubo_matrix(G: nx.Graph) -> np.ndarray:
    """
    Construye la matriz Q del problema QUBO correspondiente al MAXCUT
    de un grafo ponderado no dirigido G (NetworkX Graph).

    Par치metros:
    - G: grafo NetworkX con pesos en las aristas (atributo 'weight')

    Retorna:
    - Q: matriz numpy de tama침o n x n, donde n es el n칰mero de nodos en G
    """
    n = G.number_of_nodes()
    Q = np.zeros((n, n))

    # G.edges(data=True) devuelve una lista de (v_j,v_i,{'weight': peso})
    for j, i, peso in G.edges(data=True):
        w = peso.get('weight', 1.0)  # peso por defecto = 1 si no est치 definido
        Q[i, i] -= w
        Q[j, j] -= w
        # i es siempre > j
        Q[i, j] += 2 * w

    return Q
```

```python
Q = maxcut_qubo_matrix(G)
print(Q)
```

```python
# Algoritmo MAXCUT por fuerza bruta
import itertools

def solve_maxcut_bruteforce(Q: np.ndarray):
    """
    Resuelve el problema MAXCUT exactamente por fuerza bruta a partir de la matriz Q del QUBO.

    Par치metros:
    - Q: matriz QUBO (numpy array de tama침o n x n)

    Retorna:
    - x_opt: vector binario 칩ptimo
    - max_cut_value: valor m치ximo del corte
    """
    n = Q.shape[0]
    max_cut_value = float('-inf')
    x_opt = None

    # Producto cartesiano {0,1}^n
    # Genera todas las tuplas (x_0,..,x_{n-1}) con x_i={0,1}
    for x in itertools.product([0, 1], repeat=n):
        x_vec = np.array(x)
        cut_value = - x_vec @ Q @ x_vec  # recordamos que la QUBO representa -MAXCUT
        if cut_value > max_cut_value:
            max_cut_value = cut_value
            x_opt = x

    return np.array(x_opt), max_cut_value
```

```python
corte, peso = solve_maxcut_bruteforce(Q)
print('Soluci칩n = ',corte)
print('Peso = ',peso)
```

## **Modelo Ising**

El [_modelo Ising_](https://en.wikipedia.org/wiki/Ising_model) es un modelo matem치tico usado en f칤sica estad칤stica para estudiar el comportamiento de materiales ferromagn칠ticos.

Este modelo parte de una malla de $n$ part칤culas y busca minimizar la siguiente funci칩n:

$$
H(\textbf{z}) = -\sum_{(i,j) \in E} J_{ij}z_iz_j -\sum_j h_j z_j
$$

donde $z_i = \{-1,1\}$ representa el esp칤n de la part칤cula $i$, $E$ el conjunto de aristas de la malla, $J_{ij}$ la energ칤a de interacci칩n entre dos part칤culas y $h_i$ la influencia de un campo magn칠tico externo.

$H(\textbf{z})$ se denomina **funci칩n Hamiltoniana** y representa la energ칤a del sistema. Resolver el problema permite obtener la configuraci칩n de espines correspondiente a un estado de m칤nima energ칤a (o estado base).

**Ejemplo**:

<center><img src="https://drive.google.com/uc?export=view&id=1t8ii_V5HKX0sCVPKqgShQQzqTZOcXVxZ" alt="Malla Ising" width="500"  /></center>


En esta malla, suponiendo que $h_i=1, \forall{i}$, el Hamiltoniano resulta:

$$
H(\textbf{z}) = z_0z_1-3z_1z_2+z_2z_3-2z_0z_4+z_1z_5-3z_2z_6-2z_3z_7\\
+z_4z_5+z_5z_6+z_6z_7+z_4z_8-3z_5z_9+z_6z_{10}-z_7z_{11}\\
+z_8z_9-2z_9z_{10}+2z_{10}z_{11}-z_0-z_1-z_2-z_3-z_4-z_5\\
-z_6-z_7-z_8-z_9-z_{10}-z_{11}
$$

con $z_i = \{-1,1\}, \forall i$

Es posible demostrar que cualquier problema QUBO puede expresarse como un modelo Ising de forma simple.

### Expresi칩n mediante operadores cu치nticos

Supongamos la siguiente malla simple:

<center><img src="https://drive.google.com/uc?export=view&id=1pHIFMBSBWWciVhG1ll_SAKibigkHOiKz" alt="Malla Ising 3 nodos" width="350"  /></center>

En este caso, suponiendo de nuevo que $h_i=1, \forall{i}$, el problema queda como sigue:

\begin{align*}
\text{Minimizar}\quad & z_1 z_0 - 3z_2z_1 \\
\text{con}\quad & z_i = \{-1,1\}, i = 0,1,2
\end{align*}

Para el operador Pauli-Z, los valores esperados en los estados $|0\rangle$ y $|1\rangle$ son:

$$
\begin{aligned}
\langle 0|Z|0\rangle &= \begin{bmatrix}1 & 0\end{bmatrix}\begin{bmatrix}1 & 0\\0 & -1\end{bmatrix}\begin{bmatrix}1 \\ 0\end{bmatrix} = 1\\
\langle 1|Z|1\rangle &= \begin{bmatrix}0 & 1\end{bmatrix}\begin{bmatrix}1 & 0\\0 & -1\end{bmatrix}\begin{bmatrix}0 \\ 1\end{bmatrix} = -1
\end{aligned}
$$

Por lo tanto, podemos expresar el problema anterior como:

\begin{align*}
\text{Minimizar}\quad & \langle x_1|Z|x_1\rangle \langle x_0|Z|x_0\rangle - 3 \langle x_2|Z|x_2\rangle \langle x_1|Z|x_1\rangle\\
\text{con}\quad & x_i = \{0,1\}, i = 0,1,2
\end{align*}

Supongamos un estado base de un sistema de 3 c칰bits, $|x\rangle = |x_2x_1x_0\rangle$ con $x_i = \{0,1\}$.

La propiedad del producto mixto nos permite escribir:

$$
\langle x_1|Z|x_1\rangle \langle x_0|Z|x_0\rangle =
\langle x_2|I|x_2\rangle \langle x_1|Z|x_1\rangle \langle x_0|Z|x_0\rangle =
\langle x_2x_1x_0|I\otimes Z \otimes Z|x_2x_1x_0\rangle = \langle x|Z_1Z_0|x\rangle
$$

donde el operador $Z_i$ act칰a sobre el c칰bit $i$.

Por lo tanto, el problema anterior se puede expresar como buscar el estado base $|x_\text{min}\rangle$ que minimize la siguiente expresi칩n:

$$
\langle x_\text{min}|Z_1Z_0|x_\text{min}\rangle - 3\langle x_\text{min}|Z_2Z_1|x_\text{min}\rangle = \langle x_\text{min}|Z_1Z_0-3Z_2Z_1|x_\text{min}\rangle
$$

#### Estados arbitrarios

Supongamos en un espacio de $n$ c칰bits un estado arbitrario $\vert \psi\rangle = \sum_{i=0}^{2^n-1} a_i\vert i\rangle$. Dado $l,m$ arbitrarios, tenemos:

$$
\langle \psi\vert Z_lZ_m\vert \psi\rangle =
\left(\sum_{i=0}^{2^n-1} a_i^*\langle i\vert\right) Z_lZ_m\left(\sum_{j=0}^{2^n-1} a_j\vert j\rangle\right) =
\sum_{i=0}^{2^n-1}\sum_{j=0}^{2^n-1} a_i^*a_j\langle i\vert Z_lZ_m\vert j\rangle
$$

Adem치s, para un estado de la base $\vert i\rangle$ se verifica:

$$
Z_jZ_k|i\rangle = \pm|i\rangle
$$

es decir, los estados de la base son autovectores de $Z$ con autovalor $\pm 1$.

Y como $\langle i\vert  j\rangle = 0, \forall i\ne j$, la expresi칩n anterior queda:

$$
\langle \psi\vert Z_lZ_m\vert \psi\rangle = \sum_{i=0}^{2^n-1} |a_i|^2\langle i\vert Z_lZ_m\vert i\rangle \ge \langle x_\text{min}|Z_lZ_m|x_\text{min}\rangle
$$

Es decir, el m칤nimo siempre se alcanza en uno de los estados base.

#### Generalizaci칩n

Los t칠rminos lineales del modelo Ising se pueden expresar como:

$$
\sum_{j} h_j \langle\psi|Z_j|\psi\rangle
$$

Por lo que resolver el problema Ising es equivalente a **encontrar el estado $\psi$ que minimiza**:

$$
-\sum_{(i,j) \in E} J_{ij}\langle\psi|Z_iZ_j|\psi\rangle -\sum_j h_j \langle\psi|Z_j|\psi\rangle
$$

**Ejemplo**: para la malla Ising de tres nodos encuentra por fuerza bruta el estado de la base $|x_\text{min}\rangle$ que minimiza:

$$
\langle x_\text{min}|Z_1Z_0-3Z_2Z_1|x_\text{min}\rangle
$$



```python
from qiskit.quantum_info import SparsePauliOp

n = 3

# Definimos el operador hamiltoniano
Z1Z0 = SparsePauliOp('IZZ')
Z2Z1 = SparsePauliOp('ZZI')

operador = Z1Z0 - 3*Z2Z1

# Otra forma
# operador = SparsePauliOp.from_list([('IZZ',1), ('ZZI',-3)])

print(operador)
```

```python
from qiskit.quantum_info import Statevector

# Creamos una lista con todos los estados base para n c칰bits

estados = [Statevector.from_int(i, dims=2**n) for i in range(2**n)]

for i in range(2**n):
  print('Estado ',i, 'valor esperado=',estados[i].expectation_value(operador).real)
```

Los estados $|001\rangle$ y $|110\rangle$ obtienen el valor esperado m칤nimo (-4).

Esos estados corresponde a $z_1=z_2=\pm 1$ y $z_0\ne z_1$

## **Equivalencia QUBO $\leftrightarrow$ Ising**

Es f치cil comprobar que es posible convertir un problema QUBO, con $x_i\in \{0,1\}$, en un problema Ising, con $x_i \in \{-1,1\}, y viceversa, con un simple cambio de variables:

$$
z_i = 1 - 2x_i =
\begin{cases}
1, \text{si } x_i = 0 \\
-1, \text{si } x_i = 1
\end{cases}

Existen otras formulaciones para problemas de optimizaci칩n, seg칰n el tipo de variables y la imposici칩n de restricciones:

Este tipo de problemas pueden expresarse como problemas QUBO con ciertas limitaciones.

### **Programaci칩n binaria lineal**

Problemas de optimizaci칩n binaria **con restricciones**:

$$
\begin{aligned}
\text{Minimizar} \quad & c_0x_0+c_1x_1+\dots +c_{n-1}x_{n-1} \\
\text{sujeto a} \quad & Ax \le b, \\
& x_i = \{0,1\}
\end{aligned}
$$

con $c_i,a_{ij},b_j \in \mathbb{Z}$.


Es posible escribir un problema de este tipo como un problema QUBO usando variables de holgura (*slack variables*) y  t칠rminos de penalizaci칩n (*penalty terms*).

Nota: tambi칠n es posible generalizar al caso $c_i,a_{ij},b_j \in \mathbb{R}$

**Ejemplo**:

$$
\begin{aligned}
\text{Minimizar} \quad & -5x_0+3x_1+-2x_2 \\
\text{sujeto a} \quad & x_0+x_2 \le 1, \\
& 3x_0-x_1+3x_2 \le 4 \\
& x_i = \{0,1\}
\end{aligned}
$$

Paso 1: Convertir las restricciones de desigualdad en igualdades usando **variables de holgura binarias**.

$$
x_0+x_2 \le 1 \Leftrightarrow  x_0+x_2+y_0 = 1, \text{ con } y_0\in\{0,1\}
$$

$$
3x_0-x_1+3x_2 \le 4 \Leftrightarrow  3x_0-x_1+3x_2 + y = 4, \text{ con } y\in [0,5]
$$

El valor de $y$ en este caso viene dado porque el valor m칤nimo de $3x_0-x_1+3x_2$ es $-1$, por lo que la igualdad se verifica con $y=5$.

Podemos expresar $y$ con variables binarias haciendo:

$$
y = y_1+2y_2+2y_3, \text{ con } y_i\in\{0,1\}
$$

De esta forma, el problema original se puede escribir como:

$$
\begin{aligned}
\text{Minimizar} \quad & -5x_0+3x_1+-2x_2 \\
\text{sujeto a} \quad & x_0+x_2+y_0-1 = 0, \\
& 3x_0-x_1+3x_2+y_1+2y_2+2y_3-4=0 \\
& x_i,y_i = \{0,1\}
\end{aligned}
$$

Las dos restricciones las podemos incluir en la funci칩n a minimizar como **t칠rminos de penalizaci칩n** para obtener una formulaci칩n QUBO:

$$
\begin{aligned}
\text{Minimizar} \quad & -5x_0+3x_1+-2x_2+P(x_0+x_2+y_0-1)^2+P(3x_0-x_1+3x_2+y_1+2y_2+2y_3-4)^2 \\
& x_i,y_i = \{0,1\}
\end{aligned}
$$

siendo $P$ una constante entera suficientemente grande para garantizar que no se escoge una soluci칩n que no cumpla las restricciones.

Por ejemplo, en este ejemplo como $-5x_0+3x_1+-2x_2 \in [-7,3]$ si se toma $
P=11$ cualquier soluci칩n que no verifique alguna de las restricciones no ser치 m칤nima.

### **Programaci칩n entera lineal**

En este caso, las variables $x_i\in \mathbb{Z}$.

Para convertirlo a QUBO, tenemos que expresar las variables enteras como binarias. As칤, si sabemos que existen un valor superior $M$ para las variables (impuesto por las restricciones), una variable entera $x_i \in \{0, 1, \dots, M\}$ se puede representar como:

$$
x_i = \sum_{k=0}^{n-1} 2^k \cdot x_{ik} \quad \text{con } x_{ik} \in \{0, 1\}
$$

con $n=\lceil \log_2 M \rceil$.

En el caso m치s general en el que $x_i \in \{N, N+1, \dots, M\}$ con $N\lt 0 \lt M$ se puede escribir:

$$

$$

con $n=\lceil \log_2 (M-N+1) \rceil$

### **Programas cuadr치ticos con restricciones cuadr치ticas**

Un programa cuadr치tico con restricciones cuadr치ticas (_Quadratically Constrained Quadratic Program_, QCQP) es un problema de optimizaci칩n con una funci칩n objetivo cuadr치tica y restricciones cuadr치ticas.

Se puede escribir como:
$$
\begin{align}
\text{minimizar} &&x^T Q x + c^T x &&\\
&& && \\
\text{sujeto a} &&Ax \leq b  &&\\

&& l_j \leq x_j \leq u_j \\

Un QCQP puede expresarse como QUBO usando los mecanismos mencionados anteriormente.
